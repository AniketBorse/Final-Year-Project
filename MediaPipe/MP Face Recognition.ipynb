```python
import cv2
import mediapipe as mp
import face_recognition

# Initialize Mediapipe Face Detection
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.5)

# Load a known face image and encode it
known_image = face_recognition.load_image_file("images/Shivam Nikam.jpg")
known_face_encoding = face_recognition.face_encodings(known_image)[0]

# Start video capture
cap = cv2.VideoCapture(0)

i = 0
counter = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    
    if counter > 100:
        i -= 1
        counter -= 100
    else:
        counter += 1

    if i < 1:
        
        # Detect faces
        results = face_detection.process(rgb_frame)

        if results.detections:
            for detection in results.detections:
                bboxC = detection.location_data.relative_bounding_box
                ih, iw, _ = frame.shape
                x, y, w, h = int(bboxC.xmin * iw), int(bboxC.ymin * ih), int(bboxC.width * iw), int(bboxC.height * ih)

                # Extract face and encode it
                face_image = rgb_frame[y:y+h, x:x+w]
                face_encodings = face_recognition.face_encodings(face_image)

                if face_encodings:
                    match = face_recognition.compare_faces([known_face_encoding], face_encodings[0], tolerance=0.6)

                    if match[0]:
                        label = "Match"
                    else:
                        label = "Unknown"

                    # Draw bounding box and label
                    cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                    cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                    print("match")
                    
            i += 1

    # Show the frame
    cv2.imshow("Face Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
