```python
import cv2
import threading

# Capture video from camera
cap = cv2.VideoCapture(0)

# Initialize frame counter and desired frame rate
frame_count = 0
frame_rate = 10  # Analyze every 10th frame

# Initialize flag to indicate if the analysis thread is running
analyze_running = False

# Define a function to perform the analysis on the 10th frame
def analyze_frame(frame):
    # Perform the analysis here
    print("analysed")

    # Set the flag to indicate that the analysis thread is no longer running
    global analyze_running
    analyze_running = False

# Start the main processing loop
while True:
    # Read a frame from the camera
    ret, frame = cap.read()

    # Check if frame was successfully captured
    if not ret:
        break

    # Increment frame counter
    frame_count += 1

    # Display the frame
    cv2.imshow('Original Frame', frame)

    # If this is the 10th frame, start a new thread to analyze the frame
    if frame_count % frame_rate == 0 and not analyze_running:
        analyze_running = True
        analyze_thread = threading.Thread(target=analyze_frame, args=(frame,))
        analyze_thread.start()

    # Check for user input to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

analyze_running = False

def analyze_frame(frame):
    # Perform the analysis here
    print("analysed")

    # Set the flag to indicate that the analysis thread is no longer running
    global analyze_running
    analyze_running = False

cap = cv2.VideoCapture(0)  # capture video from default camera (index 0)
frame_count = 0  # initialize frame counter

while True:
    ret, frame = cap.read()  # read a frame from the video capture

    if not ret:
        break  # if frame not read successfully, break the loop

    frame_count += 1  # increment frame counter

    if frame_count == 15 and not analyze_running:  # process only the 15th frame
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        cv2.imshow('Grayscale frame', gray_frame)
        analyze_running = True
        analyze_thread = threading.Thread(target=analyze_frame, args=(frame,))
        analyze_thread.start()
        analyze_thread.join()
        
        
    if frame_count == 30:  # reset the frame counter after 30 frames
        frame_count = 0

    if cv2.waitKey(1) == ord('q'):  # break the loop if 'q' is pressed
        break

cap.release()  # release the video capture
cv2.destroyAllWindows()  # destroy all windows
```
